```python
from inference import Inference_model
cls = Inference_model(path_to_model='checkpoints/latest.pth') # weights.pth
```

```python
import mmcv
import matplotlib.pyplot as plt
img = mmcv.imread('asbest/camera_pits/images/5,7-0.0-637370557933275325.bmp', flag='grayscale')
boxes, *_ = cls.predict(img)
plt.figure(figsize=(10,10))
plt.imshow(img, cmap=plt.cm.gray)
for box in boxes:
    if box[4] > 0.7:
        rect = plt.Rectangle((box[0], box[1]), box[2] - box[0], box[3] - box[1], color ='r', fill=False)
        plt.gca().add_patch(rect)
```
